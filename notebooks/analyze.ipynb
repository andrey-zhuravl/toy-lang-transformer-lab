# Analysis demo
Visualising attention heads and embeddings from a trained model.

```python
from pathlib import Path
import yaml
from src.analyzer import load_model, extract_embeddings, reduce_embeddings, plot_embeddings
from src.vocab import build_vocabulary_from_files

checkpoint_path = sorted(Path('checkpoints').glob('model_epoch_*.pt'))[-1]
model, config = load_model(checkpoint_path)
vocab = build_vocabulary_from_files(['data/dicts/base.txt'])
embeddings = extract_embeddings(model)
method = config.get('analysis', {}).get('embedding_reduce', 'tsne')
embeddings_2d = reduce_embeddings(embeddings, method=method)
plot_embeddings(embeddings_2d, vocab, save_path='reports/embeddings.png')
```
